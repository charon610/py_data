```python
# 데이터 테이블 작성
import pandas as pd
import numpy as np
df = pd.DataFrame({'id':['1','2','3','4','5'],
                  'kuk':[90, 70, 40, 80, 50],
                  'eng':[100, 80, 60, 70, 30],
                  'mat':[80, 90, 50, 90, 70]})
df
```

```python
# 수학점수가 90점 이상인 학생들만 출력

print(df['mat'] >=90)
```

```python
# 총점 필드 추가
df['총점'] = (df['kuk']+df['eng']+df['mat'])
df
```

```python
# 평균 필드 추가 
df['평균'] = (df['kuk']+df['eng']+df['mat'])/3

df
```

```python
# 평균 내림차순
df.sort_values('평균', ascending=False)
```

```python
# 평균 점수가 60점 미만이면 결측치를 입력
df.loc[(df.평균 <60), '평균']=np.NaN

df
```
